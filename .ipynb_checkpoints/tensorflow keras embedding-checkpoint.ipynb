```python
import nltk
import random
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import TensorBoard
from tensorflow.keras.preprocessing.text import one_hot,Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Flatten ,Embedding,Input, Activation, Dropout,GlobalAveragePooling1D
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import regularizers
from tensorflow.keras import optimizers
from data_utils import data_preprocess
from sklearn.model_selection import train_test_split
import time
```

```python
df = data_preprocess('train.json')
train_df = df
```

# Get all possible ingredients

```python
all_ingredients = []
max_doc_len = -1
for i in range(train_df.shape[0]):
    token = nltk.word_tokenize(train_df['ingredients'][i])
    if(len(token)>max_doc_len):
        max_doc_len = len(token)# Get all possible ingredients
    all_ingredients.extend(token)
all_ingredients=set(all_ingredients)
```

# Convert ingredients and labels to one hot

```python
dict_word_train=[]
for item in train_df['ingredients']:
    dict_word_train.append(item)
    
vocab_size=len(all_ingredients) #Total number of ingredients
encod_corp=[]

for i,doc in enumerate(dict_word_train):
    encod_corp.append(one_hot(doc,vocab_size)) #Unique integer assignment to each ingredient
    
X_train=pad_sequences(encod_corp,maxlen=max_doc_len,padding='post',value=0.0) #padding zeros to each recipe
c = train_df['cuisine'].astype('category')
Y_train=to_categorical(c.cat.codes) #one_hot_encoding labels
cuisine_code = dict(enumerate(c.cat.categories))
```

```python
x_train, x_val, y_train, y_val = train_test_split(X_train, Y_train, test_size=0.3)
x_val, x_test, y_val, y_test = train_test_split(x_val, y_val, test_size=0.33)
```

# Keras model

```python
# Get all possible ingredientsno_docs = len(dict_word_train)
lr = 0.003
adm = optimizers.Adam(lr=lr)

model = Sequential()
model.add(Embedding(vocab_size, 16))
model.add(GlobalAveragePooling1D())
model.add(Dense(60, activation=tf.keras.activations.relu, kernel_regularizer=regularizers.l2(0.0001),))
model.add(Dense(20, activation=tf.keras.activations.sigmoid, kernel_regularizer=regularizers.l2(0.0001),))
model.compile(optimizer=adm, loss='categorical_crossentropy', metrics=['acc'])
model.summary()
```

```python
NAME = "lr={}_{}".format(lr,int(time.time()))
tensorboard = TensorBoard(log_dir="logs/{}".format(NAME), update_freq='epoch')
```

```python
history = model.fit(x_train, y_train, batch_size=32, epochs=10, verbose=0,validation_data=(x_val, y_val),callbacks=[tensorboard])
```

```python
test_result = model.evaluate(x_test,y_test)
```

```python
print("loss = {}, accuracy = {}".format(test_result[0],test_result[1]))
```

# Prediction

```python
result = model.predict(X_train)
result_code=[]
for x in result:
    result_code.append(np.argmax(x))
result_label = []
for i in result_code:
    result_label.append(cuisine_code[i])
Prediction = pd.DataFrame({'Prediction':result_label})
prediction = pd.concat([train_df,Prediction],axis=1,sort=False)

prediction
```

# Loss and accuracy graphs

```python
import matplotlib.pyplot as plt

history_dict=history.history
acc = history_dict['acc']
val_acc = history_dict['val_acc']
loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(acc) + 1)
# "bo" is for "blue dot"
plt.plot(epochs, loss, 'bo', label='Training loss')
# b is for "solid blue line"
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
plt.clf() 

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```
