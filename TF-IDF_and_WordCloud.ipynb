# Term Frequency - Inverse Data Frequency

```python
import nltk
import random
import numpy as np
import pandas as pd
from data_utils import data_preprocess
from tensorflow.keras.preprocessing.text import one_hot,Tokenizer
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from tqdm import tqdm
from scipy import sparse
from wordcloud import WordCloud,ImageColorGenerator
import matplotlib.pyplot as plt
import operator
from nltk.corpus import stopwords
from PIL import Image
```

```python
df = data_preprocess('train.json')
train_df = df
```

# Create count matrix

```python
dict_word_train=[]
for item in train_df['ingredients']:
    dict_word_train.append(item)
```

```python
all_ingredients = []
max_doc_len = -1
for i in range(train_df.shape[0]):
    token = nltk.word_tokenize(train_df['ingredients'][i])
    if(len(token)>max_doc_len):
        max_doc_len = len(token)
    all_ingredients.extend(token)
all_ingredients=list(set(all_ingredients))
```

```python
Max_features = len(all_ingredients)
vectorizer = CountVectorizer(max_df=0.1,max_features=Max_features)  
x_train = vectorizer.fit_transform(dict_word_train).toarray()
c = train_df['cuisine'].astype('category')
y_train=c.cat.codes
cuisine_code = dict(enumerate(c.cat.categories))
cuisine_names = list(cuisine_code.values())
```

```python
termCountMatrix = np.zeros((len(cuisine_code),x_train.shape[1]))
for j in tqdm(range(train_df.shape[0])):
    cuisine_encod = cuisine_names.index(train_df['cuisine'].iloc[j])
    for k,ingre in enumerate(x_train[j]):
        if(ingre == 1):
            termCountMatrix[cuisine_encod][k]+=1
```

# Get tfidf values

```python
countsMatrix = sparse.csr_matrix(termCountMatrix)
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(countsMatrix) 
tfidf = tfidf.toarray()
```

# Get dictionary of cuisine with corresponding ingredients and tfidf values

```python
a=np.ones((x_train.shape[1],))
g=vectorizer.inverse_transform(a)[0]
cuisine_dict = dict(enumerate(g))
```

```python
cuisine_1000={}
for x in range(20):
    cuisine_ingre = dict(enumerate(tfidf[x]))
    sorted_x = sorted(cuisine_ingre.items(), key=operator.itemgetter(1))
    sorted_1000 = sorted_x[-1000:]
    for i,ing in enumerate(sorted_1000):
        sorted_1000[i]=(cuisine_dict[ing[0]],ing[1])
    cuisine_1000[cuisine_names[x]] = dict(sorted_1000)
```

# Wordclouds

```python
#Word cloud for Brazil
cloud= np.array(Image.open("images/dcloud.png"))
color = Image.open("images/brazil.png").convert("RGB")
color = color.resize((600,316))
color = np.array(color)

wordcloud_brazil = WordCloud(background_color="white", mode="RGBA", mask=cloud)
wordcloud_brazil = wordcloud_brazil.fit_words(cuisine_1000['brazilian'])

image_colors = ImageColorGenerator(color)
plt.figure(figsize=[12,12])
plt.imshow(wordcloud_brazil.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```
